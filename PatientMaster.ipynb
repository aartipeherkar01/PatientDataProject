84.	Detect BMI outliers using IQR
📝 Note:
o	Lower bound = Q1 − 1.5×IQR
o	Upper bound = Q3 + 1.5×IQR

````python
Q1 = pdf['BMI'].quantile(0.25)
Q3 = pdf['BMI'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = pdf[(pdf['BMI'] < lower_bound) | (pdf['BMI'] > upper_bound)]
print(outliers)
````

24. Plot line chart of patient registrations per month
📝 Note: Seasonal registration patterns.


> Add blockquote



````python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to datetime (important)
pdf["Registration_Date"] = pd.to_datetime(pdf["Registration_Date"], dayfirst=True, errors='coerce')

# Drop invalid dates
pdf = pdf.dropna(subset=["Registration_Date"])

# Extract Month-Year
pdf["Month"] = pdf["Registration_Date"].dt.to_period("M")

# Group by Month
monthly_reg = pdf.groupby("Month").size().sort_index()

# Convert index to string for plotting
monthly_reg.index = monthly_reg.index.astype(str)

# Plot line chart
plt.figure(figsize=(10,5))
monthly_reg.plot(marker='o', linewidth=2)

# Styling
plt.title("Patient Registrations Per Month", fontsize=14, fontweight="bold")
plt.xlabel("Month", fontweight="bold")
plt.ylabel("Number of Patients", fontweight="bold")

plt.grid(True, linestyle="--", alpha=0.5)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````
